```python
import pandas as pd
import numpy as np
```

```python
coverage = pd.read_csv('../../data/treemmer/autocycler_full_coverage_summarized.csv', index_col = 0)
strains = pd.read_csv('../../data/treemmer/nanopore_genomes.final_tree.tre_trimmed_list_X_10', header = None)[0].values
strains = np.append(strains, 'EEE029')

coverage_strains = coverage[coverage['strain'].isin(strains)]
coverage_strains.reset_index(inplace = True, drop = True)
```

```python
coverage_strains.loc[:,'25x'] = 25/coverage_strains['coverage'].copy().values
coverage_strains.loc[:,'50x'] = 50/coverage_strains['coverage'].copy().values
coverage_strains.loc[:,'75x'] = 75/coverage_strains['coverage'].copy().values
coverage_strains.loc[:,'100x'] = 100/coverage_strains['coverage'].copy().values
coverage_strains.loc[:,'125x'] = 125/coverage_strains['coverage'].copy().values
coverage_strains.to_csv('subsample_reads_snakemake/frac_for_read_depth.csv')
```
