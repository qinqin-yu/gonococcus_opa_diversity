```python
from Bio import SeqIO
import glob
import os
```

```python
assemblies_path = "../../../../opa_diversity_snakemake/input_data/complete_genome_assemblies/grad_lab/"
chromosome_list_path = "../data/assembly_chromosome_list_files/"
os.makedirs(chromosome_list_path, exist_ok=True)

filenames = glob.glob(assemblies_path + '*.fa')
for filename in filenames:
    # Get isolate name
    isolate = filename.split('/')[-1][:-3]
    
    # Load records from fasta file
    records = list(SeqIO.parse(filename, "fasta"))
    
    # Open new file for writing chromosome list file
    with open(chromosome_list_path + isolate + ".txt", "w") as file:
        
        # Loop through records (chromosome and plasmids)
        for record in records:
            
            # Get names and lengths of each entry in fasta file
            name = record.id
            description = record.description.split(' ')
            length = int(description[1].split('=')[1])
            
            # Check if it's the chromosome
            if length>2*10**6:
                chromosome_type = 'Chromosome'
            else:
                chromosome_type = 'Plasmid'

            # Write to chromosome file
            file.write(name + '\t' + name + '\t' + chromosome_type + '\n')
```
