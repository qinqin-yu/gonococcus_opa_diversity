```python
import pandas as pd
import os
```

```python
# Create folder for manifest files if it doesn't already exist
os.makedirs('../data/manifest_files', exist_ok=True)

# Read in overall manifest file
df = pd.read_csv('../data/ena_reads_submission_manifest_combined.tsv', sep = '\t')

# Get column values
columns = df.columns

# Iterate through isolate and create a separate manifest for each file (field name + \t + field value + \n)
for i, row in df.iterrows():
    name = row['NAME']
    with open("../data/manifest_files/" + name + ".txt", "w") as file:
        for column in columns:
            file.write(column + '\t' + row[column] + '\n')
```
