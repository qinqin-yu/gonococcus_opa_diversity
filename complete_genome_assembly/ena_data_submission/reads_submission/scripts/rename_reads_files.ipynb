```python
import pandas as pd
import os
import glob
import shutil
import numpy as np
```

```python
# Create directory where the renamed read files will be saved
os.makedirs('/n/holylfs05/LABS/grad_lab/Users/qinqinyu/datasets/opa_manuscript_ncbi_submission/nanopore_sequencing_reads/bams/', exist_ok=True)

# Get names of all complete genomes sequenced in the Grad lab used in the study
complete_genomes_metadata = pd.read_csv('/n/grad_lab2/Users/qinqinyu/projects/opa_project/opa_diversity_snakemake/input_data/whole_genome_metadata/grad_lab_complete_genomes_metadata.csv', index_col = 0)
complete_genomes = complete_genomes_metadata['wgs_id'].values

# Loop through folders
folders = ['230525', '230621', '230802', '230913', '231031', '240821', '240829_gc_qy_representative_genomes_batch2', '241009_gc_sb_qy_collab_representative_genomes_batch3']
dest = '/n/holylfs05/LABS/grad_lab/Users/qinqinyu/datasets/opa_manuscript_ncbi_submission/nanopore_sequencing_reads/bams/'

for folder in folders:
    # Get the correspondence between sample name and barcode number
    samples = pd.read_csv('/n/grad_lab2/Lab/gonococcus/nanopore_sequencing/' + folder + '/samples.txt', sep = '\t', index_col = False)
    
    # Loop through the read files with the "barcode" names
    read_files = np.sort(glob.glob('/n/grad_lab2/Lab/gonococcus/nanopore_sequencing/' + folder + '/demultiplexed_reads/*barcode*.bam'))
    for read_file in read_files:
        # Get the barcode name from the read file
        barcode = read_file.split('SQK-NBD114-24_')[1][:-4]
        # Get the corresponding wgs id
        wgs_ids = samples['wgs_id'][samples['barcode']==barcode].values
        
        # Check if the corresponding wgs id wasn't found, then use just the barcode number (instead of 'barcode##')
        if len(wgs_ids) == 0:
            wgs_ids = samples['wgs_id'][samples['barcode']==int(barcode.split('barcode')[1])].values
        
        # If there was only one wgs id corresponding to this barcode (what we would expect)
        if len(wgs_ids) == 1:
            # Get that wgs id
            wgs_id = wgs_ids[0]
            # Check if the wgs id is in the list of complete genomes that we sequenced (since some of the nanopore data is of lab-derived isolates)
            print(wgs_id)
            if wgs_id in complete_genomes:
                # Rename JJJ025 (unpublished isolate from SSS project) to EEE024 (identical published isolate - what is used in the manuscript)
                if wgs_id == 'JJJ025':
                    wgs_id = 'EEE024'
                # If the read file has not previously been copied over, then copy it over
                if not os.path.exists(dest + wgs_id + '.bam'):
                    shutil.copy(read_file, dest + wgs_id + '.bam') 
                    print('copied')
        
        # If there was more than one wgs id corresponding to this barcode, then throw a message
        elif len(wgs_ids)>1: 
            print('More than one wgs id found for this barcode')
```
