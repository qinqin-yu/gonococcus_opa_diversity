```python
import egglib
import glob
import pandas as pd
```

```python
# Define operator to calculate segregating sites

cs = egglib.stats.ComputeStats()
cs.add_stats('S')
alphabet = egglib.Alphabet('char', ['a', 't', 'c', 'g', '-', 'N'], [], case_insensitive = True)
```

```python
# Calculate the number of segregating sites for all opa alignments
filenames = glob.glob('../../data/kit14/opa_sequences_aligned/*.aln')

filename = '../../data/kit14/opa_sequences_aligned/DGI_65_opa_3.aln'
opa_names = []
num_different = []
for filename in filenames:
    opa_name = filename.split('/')[-1][:-4]
    aln1 = egglib.io.from_fasta(filename, labels=True, alphabet = alphabet)#egglib.alphabets.DNA)
    struct = egglib.struct_from_labels(aln1, lvl_pop=0, lvl_indiv=1)
    stats = cs.process_align(aln1, max_missing = 1)
    
    opa_names.append(opa_name)
    num_different.append(stats['S'])
    
df = pd.DataFrame({'opa_name':opa_names, 'segregating_sites':num_different})
```

```python
# These are the opas with non-zero segregating sites
df[df['segregating_sites']>0]
```

```python
# After inspecting the alignments manually, these are all changes in the repeat regions 
# in the assemblies where the parent was polished by short reads from the evolved strain with polypolish AND pypolca.
# Polishing with polypolish alone and short reads from the evolved strain did not lead to these changes.
# Additionally, the long read only assemblies had the same sequence as the short-read polished assemblies.
```
