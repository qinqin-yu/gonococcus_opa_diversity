```python
import pandas as pd
from Bio import SeqIO
import os
```

```python
# Collates the all of the same opa gene from different assemblies into one fasta file
```

```python
# Do for parent strains

strains = ['DGI_65', 'EEE029']
for strain in strains:
    
    # Initialize dictionary
    records = list(SeqIO.parse("../../data/kit14/opa_sequences/" + strain + ".flye.fa", "fasta"))
    collated_dict = {}
    for record in records:
        opa_number = record.id.split('_opa_')[1]
        collated_dict['opa_' + opa_number] = [record]

    # Add sequence records to dictionary
    filename_endings = ['.medaka.fa',
                        '.polypolish.fa',
                        '.polypolish.pypolca.fa',
                        '_parent_with_a.polypolish.fa',
                        '_parent_with_b.polypolish.fa',
                        '_parent_with_a.polypolish.pypolca.fa',
                        '_parent_with_b.polypolish.pypolca.fa']

    for filename_ending in filename_endings:
        records = list(SeqIO.parse('../../data/kit14/opa_sequences/' + strain + filename_ending, "fasta"))
        for record in records:
            opa_number = record.id.split('_opa_')[1]
            collated_dict['opa_' + opa_number].append(record)  
    
    # Save outputs
    os.makedirs('../../data/kit14/opa_sequences_collated', exist_ok = True)
    for opa_number in collated_dict:
        SeqIO.write(collated_dict[opa_number], "../../data/kit14/opa_sequences_collated/" + strain + '_' + opa_number + '.fa', "fasta")
```

```python
# Do for evolved strains

strains = ['DGI_65a_II', 'DGI_65b_IV', 'EEE029a_VI', 'EEE029b_VI']
for strain in strains:
    
    # Initialize dictionary
    records = list(SeqIO.parse("../../data/kit14/opa_sequences/" + strain + ".flye.fa", "fasta"))
    collated_dict = {}
    for record in records:
        opa_number = record.id.split('_opa_')[1]
        collated_dict['opa_' + opa_number] = [record]

    # Add sequence records to dictionary
    filename_endings = ['.medaka.fa',
                        '.polypolish.fa',
                        '.polypolish.pypolca.fa']

    for filename_ending in filename_endings:
        records = list(SeqIO.parse('../../data/kit14/opa_sequences/' + strain + filename_ending, "fasta"))
        for record in records:
            opa_number = record.id.split('_opa_')[1]
            collated_dict['opa_' + opa_number].append(record)  
    
    # Save outputs
    os.makedirs('../../data/kit14/opa_sequences_collated', exist_ok = True)
    for opa_number in collated_dict:
        SeqIO.write(collated_dict[opa_number], "../../data/kit14/opa_sequences_collated/" + strain + '_' + opa_number + '.fa', "fasta")
```
