```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Arial"
```

```python
# Example data
data = {
    "Coverage": ["21x", "42x", "84x", "140x"] * 14,
    "Changes": [1, 0, 0, 0,   # Method 1
                       1, 0, 0, 0,    # Method 2
                       0, 0, 0, 0,   # Method 3
                       0, 0, 0, 0,   # Method 4
                       0, 0, 0, 0,    # Method 5
                       0, 0, 0, 0,   # Method 6
                       0, 0, 0, 0,    # Method 7
                       1, 0, 0, 0,   # Method 1
                       1, 0, 0, 0,    # Method 2
                       0, 0, 0, 0,   # Method 3
                       0, 0, 0, 0,   # Method 4
                       0, 0, 0, 0,    # Method 5
                       0, 0, 0, 0,   # Method 6
                       0, 0, 0, 0],    # Method 7
    "Method": ((["1 (Flye)"] * 4 +
               ["2 (Flye + Medaka)"] * 4 +
               ["3 (Flye + Medaka + Polypolish)"] * 4 +
               ["4 (Flye + Medaka + Polypolish + Pypolca)"] * 4 +
               ["5 (Autocycler)"] * 4 +
               ["6 (Autocycler + Polypolish)"] * 4 +
               ["7 (Autocycler + Polypolish + Pypolca)"] * 4)*2)
}

df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(4.5, 3.5))
sns.stripplot(data=df, x="Coverage", y="Changes", hue="Method", dodge=True, jitter = 0.2, palette="Set2")

plt.yticks([0, 1])
plt.ylabel("Number of changes in $opa$ genes")
plt.xlabel("Approximate coverage")
plt.legend(title="Method")
plt.tight_layout()
plt.savefig("../../../figures/compare_methods_read_subset.png", dpi = 300)
plt.savefig("../../../figures/compare_methods_read_subset.pdf")
plt.show()
```
