```python
import pandas as pd
from Bio import SeqIO
import os
```

```python
# Collates the all of the same opa gene from different assemblies into one fasta file
```

```python
# Do for parent strains

# strains = ['DGI_65_15', 
#            'DGI_65_30', 
#            'DGI_65_60',
#            'DGI_65a_II_15',
#            'DGI_65a_II_30',
#            'DGI_65a_II_60',
#            'DGI_65b_IV_15',
#            'DGI_65b_IV_30',
#            'DGI_65b_IV_60',
#            'EEE029_15', 
#            'EEE029_30', 
#            'EEE029_60',
#            'EEE029a_VI_15',
#            'EEE029a_VI_30',
#            'EEE029a_VI_60',
#            'EEE029b_VI_15',
#            'EEE029b_VI_30',
#            'EEE029b_VI_60']

strains = ['DGI_65',
           'DGI_65a_II',
           'DGI_65b_IV',
           'EEE029',
           'EEE029a_VI',
           'EEE029b_VI']

fracs = ['15', '30', '60']

filename_endings = ['.flye.fa',
                    '.medaka.fa',
                    '.polypolish.fa',
                    '.polypolish.pypolca.fa',
                    '.autocycler.fa',
                    '.autocycler.polypolish.fa',
                    '.autocycler.polypolish.pypolca.fa']

for strain in strains:
    
    # Initialize dictionary
    records = list(SeqIO.parse("../../../data/kit14/subsampled_nanopore_reads/opa_sequences/" + strain + "_" + fracs[0] + ".flye.fa", "fasta"))
    collated_dict = {}
    for record in records:
        opa_number = record.id.split('_opa_')[1]
        collated_dict['opa_' + opa_number] = []

    # Add sequence records to dictionary

    for frac in fracs:
        for filename_ending in filename_endings:
            if os.path.exists('../../../data/kit14/subsampled_nanopore_reads/opa_sequences/' + strain + '_' + frac + filename_ending):
                records = list(SeqIO.parse('../../../data/kit14/subsampled_nanopore_reads/opa_sequences/' + strain + '_' + frac + filename_ending, "fasta"))
                for record in records:
                    opa_number = record.id.split('_opa_')[1]
                    collated_dict['opa_' + opa_number].append(record)  
    
    # Save outputs
    os.makedirs('../../../data/kit14/subsampled_nanopore_reads/opa_sequences_collated', exist_ok = True)
    for opa_number in collated_dict:
        SeqIO.write(collated_dict[opa_number], "../../../data/kit14/subsampled_nanopore_reads/opa_sequences_collated/" + strain + '_' + opa_number + '.fa', "fasta")
```
