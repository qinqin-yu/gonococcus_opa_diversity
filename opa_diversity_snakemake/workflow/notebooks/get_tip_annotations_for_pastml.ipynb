```python
import pandas as pd
import numpy as np
```

```python
# Load tip state information
metadata = pd.read_csv('../../results/opa_metadata_locus.csv', index_col = 0)
opa_lineages = pd.read_csv('../../results/mcl/mcl_clusters.csv', index_col = 0)
metadata = metadata.merge(opa_lineages, on = 'id')

# Convert all tip states to strings (needed to specify colors for plotting pastml tree visualizations)
metadata['sv_cluster'] = metadata['sv_cluster'].astype('str')
metadata['hv1_cluster'] = metadata['hv1_cluster'].astype('str')
metadata['hv2_cluster'] = metadata['hv2_cluster'].astype('str')

# Save tip information into dataframe with each column corresponding to a different type of state (i.e. opaA_hv1)

opa_loci = np.unique(metadata['locus'].dropna())
tip_annotations = pd.DataFrame({'strain':np.unique(metadata['strain'])})

for locus in opa_loci:
    tip_annotations = tip_annotations.merge(metadata[metadata['locus']==locus][['strain', 'sv_cluster']].rename({'sv_cluster':'opa' + locus + '_sv'}, axis = 'columns'), on = 'strain', how = 'left')
    tip_annotations = tip_annotations.merge(metadata[metadata['locus']==locus][['strain', 'hv1_cluster']].rename({'hv1_cluster':'opa' + locus + '_hv1'}, axis = 'columns'), on = 'strain', how = 'left') 
    tip_annotations = tip_annotations.merge(metadata[metadata['locus']==locus][['strain', 'hv2_cluster']].rename({'hv2_cluster':'opa' + locus + '_hv2'}, axis = 'columns'), on = 'strain', how = 'left')
    
# Save file
tip_annotations.to_csv('../../results/pastml/tip_annotations.txt', sep = '\t', index = False)
```
