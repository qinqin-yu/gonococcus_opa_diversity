```python
import pandas as pd
import numpy as np
import glob
```

```python
# Get wgs_ids for isolates sequenced in our lab with nanopore
filenames = glob.glob('../../input_data/complete_genome_assemblies/grad_lab/*.fa')
wgs_ids = []
for filename in filenames:
    wgs_ids.append(filename.split('/')[-1].split('.fa')[0])
wgs_ids_df = pd.DataFrame({'wgs_id':wgs_ids})

ng_metadata = pd.read_csv('/n/holylfs05/LABS/grad_lab/Lab/repos/gc_genomics/metadata/Ng-Combined-Metadata.txt', sep = '\t')

grad_lab_comp_gen_metadata = wgs_ids_df.merge(ng_metadata, how = 'left', on = 'wgs_id')
grad_lab_comp_gen_metadata[['assembly', 'bioproject', 'release_date', 'notes']] = np.nan
grad_lab_comp_gen_metadata['long_read_sequencer'] = 'Oxford Nanopore MinION'
grad_lab_comp_gen_metadata['assembly_method'] = 'Autocycler v 0.2.1'
grad_lab_comp_gen_metadata['source'] = 'this_study'

grad_lab_comp_gen_metadata_short = grad_lab_comp_gen_metadata[['wgs_id', 'accession', 'names', 'date', 'continent', 'country', 'region', 'assembly', 'bioproject', 'release_date', 'long_read_sequencer', 'assembly_method', 'notes', 'source', 'reference']]
grad_lab_comp_gen_metadata_short.rename({'date':'biosample_collection_date', 'reference':'original_reference'}, axis = 'columns', inplace = True)
grad_lab_comp_gen_metadata_short.sort_values('wgs_id', inplace = True, ignore_index = True)
grad_lab_comp_gen_metadata_short.to_csv('../../input_data/whole_genome_metadata/grad_lab_complete_genomes_metadata.csv')
```

```python
# Get representative genomes

filenames = glob.glob('../../input_data/representative_genome_pseudogenomes/*.fasta')
wgs_ids = []
for filename in filenames:
    wgs_ids.append(filename.split('/')[-1].split('_pseudogenome.fa')[0])
wgs_ids_df = pd.DataFrame({'wgs_id':wgs_ids})

metadata = pd.read_csv('../../input_data/lab_strains_hybrid_genomes_metadata.csv', index_col = 0)

representative_metadata = wgs_ids_df.merge(metadata, how = 'left', on = 'wgs_id')

representative_metadata['source'] = 'representative'
representative_metadata = representative_metadata[['wgs_id', 'accession', 'names', 'date', 'continent', 'country', 'region', 'source', 'reference']]
representative_metadata.rename({'date':'biosample_collection_date'}, axis = 'columns', inplace = True)
representative_metadata.sort_values('wgs_id', inplace = True, ignore_index = True)
representative_metadata.to_csv('../../input_data/whole_genome_metadata/representative_genomes_metadata.csv')
```
