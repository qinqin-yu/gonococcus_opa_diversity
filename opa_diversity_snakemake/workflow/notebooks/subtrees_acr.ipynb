```python
from pastml.acr import pastml_pipeline
import pandas as pd
import os
import numpy as np
```

```python
# Runs PastML ancestral state reconstruction on all opa loci and regions
# Uses a time-scaled phylogeny of high-quality Ng complete genomes
# Gives the rate of change of the states over time

subtrees = [4]#[3, 4, 6]

# Path to the table containing tip/node annotations, in csv or tab format
data = "../../results/pastml/tip_annotations.txt"
# Load in all character names (i.e. opaA_sv)
tip_annotations = pd.read_csv(data, sep = '\t')

for subtree in subtrees:
    
    # Make directories for saving outputs
    os.makedirs('../../results/subtrees/pastml/subtree_' + str(subtree) + '/map_visualizations', exist_ok=True) 
    os.makedirs('../../results/subtrees/pastml/subtree_' + str(subtree) + '/tree_visualizations', exist_ok=True) 
    os.makedirs('../../results/subtrees/pastml/subtree_' + str(subtree) + '/outputs', exist_ok=True) 

    # Get the characters to use (since some subtrees have isolates that are all missing a certain opa locus)
    isolates = pd.read_csv('../../results/subtrees/isolates/subtree_' + str(subtree) + '_isolates.txt', header = None)[0].values
    tip_annotations_subtree = tip_annotations[tip_annotations['strain'].isin(isolates)]
    tip_annotations_subtree.dropna(axis = 'columns', how = 'all', inplace = True)
    all_columns = tip_annotations_subtree.columns[1:]

    # Loop through all characters 
    for column in all_columns:
        
        # Get whether it's sv, hv1, or hv2
        region = column.split('_')[1]
        
        # Path to the tree in newick format
        tree = "../../results/subtrees/beast/subtree_" + str(subtree) + "_recomb_masked_with_dates-subtree_" + str(subtree) + "_recomb_masked_with_dates.MCC.nwk"

        # Columns present in the annotation table,
        # for which we want to reconstruct ancestral states
        # (for Albanian data we only have one column, but multiple columns are also allowed)
        columns = [column]

        # Path to the output compressed map visualisation
        html_compressed = "../../results/subtrees/pastml/subtree_" + str(subtree) + "/map_visualizations/" + column + "_map.html"

        # (Optional) path to the folder where pastml parameter, named tree
        # and marginal probability (for marginal ML methods (pastml.ml.MPPA, pastml.ml.MAP) only) files are to be stored.
        # Default is <path_to_input_file>/<input_file_name>_pastml. If the folder does not exist, it will be created.
        work_dir = '../../results/subtrees/pastml/subtree_' + str(subtree) + '/outputs'
        
        # (Optional) path to the output tree visualisation
        html = "../../results/subtrees/pastml/subtree_" + str(subtree) + "/tree_visualizations/" + column + "_tree.html"

        # (Optional) Get colour dictionary (in format {column:{state:color}}. Note that the states must be strings)
        colours_df = pd.read_csv('../../results/pastml/visualization_colors/' + region + '_colors.txt', sep = '\t', index_col = 0)
        colours_df.index = colours_df.index.astype('str') # convert state names to strings
        colours_dict = colours_df.to_dict()['colour'] # get in dictionary form
        colours_dict = {column:colours_dict}
            
        # Run PastML
        pastml_pipeline(data=data, data_sep='\t', columns=columns, name_column=column, tree=tree,
                        html_compressed=html_compressed, html=html, work_dir = work_dir, colours=colours_dict, verbose=True)
```
