```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clusters = pd.read_csv('../../results/mcl/mcl_clusters.csv', index_col = 0)
```

```python
fig, ax = plt.subplots(3,1, figsize = (5,6))
sns.countplot(data = clusters, x = 'sv_cluster', ax = ax[0], color = '#ad6a6c')
sns.countplot(data = clusters, x = 'hv1_cluster', ax = ax[1], color = '#4a5899')
sns.countplot(data = clusters, x = 'hv2_cluster', ax = ax[2], color = '#5fb49c')

ax[0].set_xlabel('Semivariable cluster type')
ax[1].set_xlabel('Hypervariable 1 cluster type')
ax[2].set_xlabel('Hypervariable 2 cluster type')

ax[0].set_ylabel('# occurrences')
ax[1].set_ylabel('# occurrences')
ax[2].set_ylabel('# occurrences')

plt.tight_layout()
plt.savefig('../../figures/clusters/cluster_num_occurrences.pdf')
plt.savefig('../../figures/clusters/cluster_num_occurrences.png', dpi = 300)
plt.tight_layout()
```
