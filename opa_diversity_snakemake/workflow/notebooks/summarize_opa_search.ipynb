```python
import pandas as pd
import numpy as np
import glob
```

```python
# Pull out the instances where the start of the coding repeat sequence was not found
filenames = glob.glob('../../results/opa_locations/*.csv')
filenames = np.sort(filenames)
for file in filenames:
    df = pd.read_csv(file, index_col = 0)
    if len(df[df['start_cr'].isna()]):
        print(file)
        print(df)
```

```python
# Get number of opa genes
metadata = pd.read_csv('../../results/opa_metadata_locus.csv', index_col = 0)
print('total number of full length opa genes: ', len(metadata))
```

```python
num_opa = pd.read_csv('../../results/summary_statistics/num_opa_genes_by_strain.csv', index_col = 0)
for num, df in num_opa.groupby('num_opa_genes'):
    print('num strains with ', num, ' opa genes:', len(df), ', ', round(100*len(df)/len(num_opa),1), '%')
```
